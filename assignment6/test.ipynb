```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/market_data.csv")
```

```python
aapl = df[df["symbol"] == "AAPL"].drop(columns=["symbol"]).set_index("timestamp")
msft = df[df["symbol"] == "MSFT"].drop(columns=["symbol"]).set_index("timestamp")
spy = df[df["symbol"] == "SPY"].drop(columns=["symbol"]).set_index("timestamp")

prices = pd.concat([aapl, msft, spy], axis=1, keys=["AAPL", "MSFT", "SPY"])
```

```python
prices.plot()
plt.tight_layout()

ax = plt.gca()
ax.set_title("Stock Prices Over Time")
ax.set_xlabel("Timestamp")
ax.set_ylabel("Price (USD)")
ax.legend(title="Symbols")
plt.show()
```

```python
aapl_returns = aapl["price"].pct_change().dropna()
msft_returns = msft["price"].pct_change().dropna()
spy_returns = spy["price"].pct_change().dropna()

returns = pd.concat(
    [aapl_returns, msft_returns, spy_returns], axis=1, keys=["AAPL", "MSFT", "SPY"]
)
```
